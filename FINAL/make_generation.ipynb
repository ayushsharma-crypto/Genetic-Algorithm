```python
import numpy as np
import os
import json
import random

def read_file(filename):
    '''
    This function will read the filename
    and return it's content.
    '''
    with open(filename,'r') as read_file:
        data=json.load(read_file)
    return data

def append_file(filename,data):
    '''
    This function will write data in the filename
    and return final content of it.
    '''
    with open(filename,'a') as write_file:
        json.dump(data, write_file, indent = 4)
        write_file.write("\n")

trace_data = read_file('trace.json')
trace=trace_data['Trace']
```

```python
FILE = 'generations/generations_10.txt'
gen = 130
```

```python
for i in range(gen):
    current_gen = trace[i]['reproduction']
    initial_population = current_gen['parent']
    after_selection = current_gen['selected']
    after_crossover = current_gen['cross']
    after_mutation = current_gen['mutated']
    
    with open(FILE,'a') as write_file:
        write_file.write("\n\nGeneration (Iteration) "+ str(i+1) + "\n")
        write_file.write("------------------------\n")
            
    with open(FILE,'a') as write_file:
        write_file.write("\nInitial population :\n\n")
        
    append_file(FILE,initial_population )
        
    with open(FILE,'a') as write_file:
        write_file.write("\nAfter Selection :\n\n")
        
    append_file(FILE,after_selection )
        
    with open(FILE,'a') as write_file:
        write_file.write("\nAfter Crossover :\n\n")
        
    append_file(FILE,after_crossover )
        
    with open(FILE,'a') as write_file:
        write_file.write("\nAfter Mutation :\n\n")
    
    append_file(FILE,after_mutation )    
```
